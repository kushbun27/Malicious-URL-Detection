```python
#1
def length_url(url):
    if len(url) < 54:
        return 0            # legitimate
    elif len(url) >= 54 and len(url) <= 75:
        return 2            # suspicious
    else:
        return 1            # phishing
df['Length']=df['Phishing Sites'].apply(length_url)
```

```python
#2
def have_at_symbol(url):
    if "@" in url:
        return 1            # phishing
    else:
        return 0
df['Having_At']=df['Phishing Sites'].apply(have_at_symbol)
```

```python
#4
def redirection(url):
    str=url
    str=str[8:]
    lst=str.split("//")
    
    if len(lst) > 1:
        return 1
    else:
        return 0
df['Redirected']=df['Phishing Sites'].apply(redirection)
```

```python
#5
#scheme://netloc/path;parameters?query#fragment
def dash(url):
    if "-" in urlparse(url).netloc:
        return 1            # phishing
    else:
        return 0            # legitimate
df['Dash']=df['Phishing Sites'].apply(dash)
```

```python
def sub_domains(url):
    if url.count(".") < 3:
        return 0            # legitimate
    elif url.count(".") == 3:
        return 2            # suspicious
    else:
        return 1            # phishing
df['Dots']=df['Phishing Sites'].apply(sub_domains)
```

```python
def shortening_service(url):
        """Tiny URL -> phishing otherwise legitimate"""
        match=re.search('bit\.ly|goo\.gl|shorte\.st|go2l\.ink|x\.co|ow\.ly|t\.co|tinyurl|tr\.im|is\.gd|cli\.gs|'
                    'yfrog\.com|migre\.me|ff\.im|tiny\.cc|url4\.eu|twit\.ac|su\.pr|twurl\.nl|snipurl\.com|'
                    'short\.to|BudURL\.com|ping\.fm|post\.ly|Just\.as|bkite\.com|snipr\.com|fic\.kr|loopt\.us|'
                    'doiop\.com|short\.ie|kl\.am|wp\.me|rubyurl\.com|om\.ly|to\.ly|bit\.do|t\.co|lnkd\.in|'
                    'db\.tt|qr\.ae|adf\.ly|goo\.gl|bitly\.com|cur\.lv|tinyurl\.com|ow\.ly|bit\.ly|ity\.im|'
                    'q\.gs|is\.gd|po\.st|bc\.vc|twitthis\.com|u\.to|j\.mp|buzurl\.com|cutt\.us|u\.bb|yourls\.org|'
                    'x\.co|prettylinkpro\.com|scrnch\.me|filoops\.info|vzturl\.com|qr\.net|1url\.com|tweez\.me|v\.gd|tr\.im|link\.zip\.net',url)
        if match:
            return 1               # phishing
        else:
            return 0               # legitimate
df['Tiny_Url']=df['Phishing Sites'].apply(shortening_service)        
```
